```python
import pandas as pd
import numpy as np
```

```python
#importing data
data=pd.read_csv("data.csv", encoding = "ISO-8859-1")
data.head()
```

```python
#renaming the name of the columns
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
```

```python
list2=['Date','label']
data.columns= list2+new_Index
data.head()
```

```python
#For easy access to each row adding a index column 
data['Index']=np.arange(len(data))
#swaping Index and Date columns
cols = list(data.columns)
a, b = cols.index('Date'), cols.index('Index')
cols[b], cols[a] = cols[a], cols[b]
data = data[cols]

# data processing such as lowering and removing the Punctuations in headlines
for index in new_Index:
    data[index]=data[index].str.lower()
    data[index]=data[index].str.replace("[^a-zA-Z]"," ",regex=True)


data.head()
```

```python
#splitting dataset into train and test using date threshold
train = data[data['Date'] < '20150101']
test = data[data['Date'] > '20141231']
print(train.shape)
print(test.shape)
```

```python
#Combining all the headlines for a particualr date in train dataset
headlines = []
for row in range(0,len(train.index)):
    
    headlines.append(' '.join(str(x) for x in train.iloc[row,2:27]))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
# bag of words
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['label'])
```

```python
#Combining all the headlines for a particualr date in test dataset
headlines_test = []
for row in range(0,len(test.index)):
    headlines_test.append(' '.join(str(x) for x in test.iloc[row,2:27]))
```

```python
test_dataset = countvector.transform(headlines_test)
predictions = randomclassifier.predict(test_dataset)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
#Confusion matrix to know the accuracy score
matrix=confusion_matrix(test['label'],predictions)
print(matrix)
score=accuracy_score(test['label'],predictions)
print(score)
report=classification_report(test['label'],predictions)
print(report)
```
